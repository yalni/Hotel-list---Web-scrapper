```python
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/List_of_hotels_in_Sri_Lanka#Colombo_District'

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

table = soup.find_all('table')[6]
```

```python
soup.find('table', class_ = 'wikitable sortable')

```

```python
#we need to get the headers first, where it is : </th>

hotel_titles = table.find_all('th')

hotel_table_titles = [title.text.strip() for title in hotel_titles ]


import pandas as pd
df = pd.DataFrame(columns = hotel_table_titles)
df
```

```python
column_data= table.find_all('tr')
```

```python
for row in column_data[1:]:
    raw_data= row.find_all('td')
    individual_row_data = [data.text.strip() for data in raw_data]
    print(individual_row_data)

    length = len(df)
    df.loc[length] = individual_row_data 
```
